```python
#Load the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the data set
df=pd.read_csv('Attrition data (1).csv')
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
#check for null values
df.isnull().sum()
```

```python
#handel the missing values by filling with 0 or mean value
df['NumCompaniesWorked'].fillna(0,inplace=True)
df['EnvironmentSatisfaction'].fillna(df['EnvironmentSatisfaction'].mean(),inplace=True)
df['JobSatisfaction'].fillna(df['JobSatisfaction'].mean(),inplace=True)
df['WorkLifeBalance'].fillna(df['WorkLifeBalance'].mean(),inplace=True)
```

```python
df['WorkLifeBalance']= df['WorkLifeBalance'].round(1)
```

```python
attrition_count = pd.DataFrame(df['Attrition'].value_counts())
attrition_count
```

```python
# pie plot for attrition % age
plt.pie(attrition_count['Attrition'] , labels = ['No' , 'Yes'] , explode = (0.2,0), autopct='%1.1f%%')
plt.title('Attrition Distribution')
plt.show()
```

```python
# number of employees by attrition status
sns.countplot(df['Attrition'], palette='Set2') 
plt.xlabel('Attrition Status')
plt.ylabel('Number of Employees')
plt.title('Count of Employees by Attrition Status')

for p in plt.gca().patches:
    plt.gca().annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                       ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```

```python
# Removing unnessesary columns
df.drop(['EmployeeCount' , 'StandardHours','Over18'] , axis = 1)
```

```python
attrition_dummies = pd.get_dummies(df['Attrition'])
attrition_dummies.head()
```

```python
df = pd.concat([df, attrition_dummies] , axis = 1)
df.head()
```

```python
df = df.drop(['Attrition' , 'No'] , axis = 1)
df.head()
```

```python
sns.barplot(x = 'Gender' , y = 'Yes', data = df)

plt.xlabel('Gender')
plt.ylabel('Number of "Attrition"')
plt.title('Number of "Attrition" by Gender')
plt.show()
```

```python
sns.barplot(x = 'Department', y = 'Yes', data = df)
plt.xlabel('Department')
plt.ylabel('Number of "Attrition"')
plt.title('Number of "Attrition" by Gender')
plt.show()
```

```python
sns.barplot(x = 'BusinessTravel', y = 'Yes', data = df)
plt.xlabel('Business Travel')
plt.ylabel('Number of "Attrition"')
plt.title('Number of "Attrition" by Business Travel')
plt.show()
```

```python
# Coorelation matrix
correlation_matrix = df.corr()
plt.figure(figsize=(20, 20))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.histplot(df['Age'], kde=True, color='skyblue')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age with KDE')
plt.axvline(df['Age'].median(), color='red', linestyle='dashed', linewidth=1.5, label='Median Age')  # Add median line
plt.axvline(df['Age'].mean(), color='orange', linestyle='dashed', linewidth=1.5, label='Mean Age')   # Add mean line
plt.legend()  
plt.grid(True)  
plt.show()
```
